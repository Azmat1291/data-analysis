```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
IMG_SIZE=224
BATCH_SIZE=32
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_datagen=ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/DAP_TEAM 4/datasets/Jellyfish/train',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)
val_generator=train_datagen.flow_from_directory(
    '/content/drive/MyDrive/DAP_TEAM 4/datasets/Jellyfish/train',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
test_datagen=ImageDataGenerator(rescale=1./255)
test_generator=test_datagen.flow_from_directory(
    '/content/drive/MyDrive/DAP_TEAM 4/datasets/Jellyfish/test',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
)
```

```python
#define the model
model=keras.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
#compile
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(train_generator,validation_data=val_generator,epochs=10)
```

```python
model.save("jelly1.h5","label.txt")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
model=load_model('/content/jelly1.h5')
test_image_path='/content/drive/MyDrive/DAP_TEAM 4/datasets/Jellyfish/test/Moon_jellyfish/05.jpg'
img=image.load_img(test_image_path,target_size=(224,224,3))
img_array=image.img_to_array(img)
img_array=np.expand_dims(img_array,axis=0)
img_array/=255
print(img_array.shape)
prediction=model.predict(img_array)
if prediction < 0.5:
  print("Prediction: Moon JellyFish (Probability:",prediction[0][0],")")
else:
  print("Prediction: other JellyFish(Probability:",prediction[0][0],")")
```
